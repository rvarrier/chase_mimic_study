```python
import pandas as pd
import numpy as np
import random
```

```python
total_vids = 112
# 6 subt * 2 start pos * 2 pred colors = 24
# * 2 variants or recollapses = 48
# * 2 conditions (chase/mimic) = 96
# + 1 wander * 2 start pos * 2 pred color * 2 variants * 2 conditions simulated = 16
# = 112 vids
```

- Let's organize these as chase first (0-48) , mimic next (48-96), wander last (96-112)
- We have 8 vids at each subtlety level for chase, mimic and wander, need to pick 6

```python
agent_subtlety = np.repeat([0,30,60,90,120,150,0,30,60,90,120,150,180,180],8)
agent_color = np.tile(np.repeat(['Gr','Bl'],4),14)
agent_startpos = np.tile(np.repeat(['L','R'],2),2*14)
motiv_type = []
motiv_type.extend(np.repeat(['Chase','Mimic',],48))
motiv_type.extend(np.repeat('Wander',16))
vid_iter_nr = np.tile(['v0','v1',],56)

len(agent_subtlety),len(agent_color),len(agent_startpos),len(agent_type), len(vid_iter_nr)
```

```python
agent_color_name = {'Gr':'Grey','Bl':'Black'}
```

```python
#  chase
vidNames = []
pos = {'L':'350','R':'850'}
for i in range(48):
    if vid_iter_nr[i] == 'v0':
        vidnumsuff = ''
        pref ='chase_'
    else:
        vidnumsuff = '_' + str(vid_iter_nr[i][1])
        pref = 'chase__'
    # chase_pred_subt0_Colorg_Posx350, 
    # chase__pred_subt0_Colorb_Posx350_2
    flname = pref + 'pred_subt' + str(agent_subtlety[i]) + '_' + agent_color_name[agent_color[i]] + '_Posx_' + pos[agent_startpos[i]] + vidnumsuff + '.webm'
    vidNames.append(flname)  
print(vidNames[:2])

# mimic
pos = {'L':'350','R':'850'}
for i in np.arange(48,96):
    if vid_iter_nr[i] == 'v0':
        pref ='mimic_flipped_'
    else:
        pref = 'mimic_flipped__'
    vidnumsuff = '_ind*' # to do!!
    # chase_pred_subt0_Colorg_Posx350, 
    # chase__pred_subt0_Colorb_Posx350_2
    flname = pref + 'pred_subt' + str(agent_subtlety[i]) + '_' + agent_color_name[agent_color[i]] + '_Posx_' + pos[agent_startpos[i]] + vidnumsuff + '.webm'
    vidNames.append(flname)
print(vidNames[47:49])

# wander
pos = {'L':'350','R':'850'}
for i in np.arange(96,112):
    if vid_iter_nr[i] == 'v0':
        vidnumsuff = ''
        pref ='wander_'
    else:
        vidnumsuff = '_' + str(vid_iter_nr[i][1])
        pref = 'wander__'
    # chase_pred_subt0_Colorg_Posx350, 
    # chase__pred_subt0_Colorb_Posx350_2
    if agent_color[i][0] == 'Bl':
        colorName = 'Black'
    else:
        colorName = "Grey"
    flname = pref + 'pred_' + colorName+ '_Posx_' + pos[agent_startpos[i]] + vidnumsuff + '.webm'
    vidNames.append(flname)
print(vidNames[95:98])
```

```python
df =  pd.DataFrame({'Vid_nr':np.arange(112),
    'Subtlety':agent_subtlety,
    'Color':agent_color,
    'StartPos': agent_startpos,
    'Motivator': motiv_type,
    'Vid_names': vidNames,
    'Iter_nr':vid_iter_nr})

ID = []
for i, rows in df.iterrows():
    id = '_'.join([str(rows['Subtlety']),str(rows['Color']),str(rows['StartPos']),str(rows['Motivator']), str(rows['Iter_nr'])])
    ID.append(id)
#df['ID'] = str(df['Vid_nr']) + str(df['Subtlety']) + str(df['Color']) + str(df['Motivator'])
df['ID'] = ID
df.head()
```
